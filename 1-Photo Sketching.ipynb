```python
import cv2 as cv
import matplotlib.pyplot as plt
```

### Oil Paint Effect

```python
# V1

img = cv.imread('images/istanbul.jpg')

oilpaint1 = cv.xphoto.oilPainting(img, 3, 4)
cv.imshow("Oil Paint V1", oilpaint1)

cv.waitKey(0)
cv.destroyAllWindows()

# Matplotlib
plt.axis("off")
plt.imshow(cv.cvtColor(oilpaint1, cv.COLOR_BGR2RGB))
plt.title("Oil Paint V1")
plt.show()
```

```python
# V2

img = cv.imread('images/istanbul.jpg')

kernel = cv.getStructuringElement(cv.MORPH_ELLIPSE, (7,7))
morph = cv.morphologyEx(img, cv.MORPH_OPEN, kernel)

oilpaint2 = cv.normalize(morph,None,10,255,cv.NORM_MINMAX)

cv.imshow("Oil Paint V2", oilpaint2)

cv.waitKey(0)
cv.destroyAllWindows()

# Matplotlib
plt.axis("off")
plt.imshow(cv.cvtColor(oilpaint2, cv.COLOR_BGR2RGB))
plt.title("Oil Paint V2")
plt.show()
```

### Watercolor Effect

```python
img = cv.imread('images/istanbul.jpg')

# sigma_s: 0-200
# sigma_r: 0-1

watercolor = cv.stylization(img, sigma_s=140, sigma_r=0.8)

cv.imshow("Watercolor",watercolor)

cv.waitKey(0)
cv.destroyAllWindows()

# Matplotlib
plt.axis("off")
plt.imshow(cv.cvtColor(watercolor, cv.COLOR_BGR2RGB))
plt.title("Watercolor")
plt.show()
```

### Cartoon Effect

```python
img = cv.imread('images/istanbul.jpg')

# Finding the Edges of Image
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY) 
gray = cv.medianBlur(gray, 7) 
edges = cv.adaptiveThreshold(gray, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 9, 10)


# Making a Cartoon of the image
color = cv.bilateralFilter(img, 1, 200, 40) 
cartoon = cv.bitwise_and(color, color, mask = edges)

#Visualize the cartoon image 
cv.imshow("Cartoon", cartoon) 

cv.waitKey(0)
cv.destroyAllWindows()

# Matplotlib
plt.axis("off")
plt.imshow(cv.cvtColor(cartoon, cv.COLOR_BGR2RGB))
plt.title("Cartoon")
plt.show()

# Source: https://towardsdatascience.com/using-opencv-to-catoonize-an-image-1211473941b6
```

### Pencil Effect

```python
img = cv.imread('images/istanbul.jpg')

# sigma_s: 0-200
# sigma_r: 0-1
# shade_factor: 0-0.1

oneColourPencil, colorfulPencil = cv.pencilSketch(img, sigma_s=150, sigma_r=0.5, shade_factor=0.02)

cv.imshow('Colorful Pencil',colorfulPencil)
cv.imshow('One Color Pencil', oneColourPencil)

cv.waitKey(0)
cv.destroyAllWindows()

# Matplotlib
plt.axis("off")
plt.imshow(cv.cvtColor(colorfulPencil, cv.COLOR_BGR2RGB))
plt.title("Colorful Pencil")
plt.show()

plt.axis("off")
plt.imshow(cv.cvtColor(oneColourPencil, cv.COLOR_BGR2RGB))
plt.title("One Color Pencil")
plt.show()
```
