```python
import cv2 as cv
import matplotlib.pyplot as plt
```

### Gray

```python
img = cv.imread('images/istanbul.jpg')

gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv.imshow('Gray', gray)

plt.imshow(gray)
cv.waitKey(0)
```

### Blur 

##### Average

This is done by convolving an image with a normalized box filter. It simply takes the average of all the pixels under the kernel area and replaces the central element. 

```python
img = cv.imread('images/istanbul.jpg')

averaging_blur = cv.blur(img,(5,5))
cv.imshow('Averaging Blur', averaging_blur)

plt.imshow(averaging_blur)
cv.waitKey(0)
```

##### Gaussian

In this method, instead of a box filter, a Gaussian kernel is used. It is done with the function, cv.GaussianBlur(). We should specify the width and height of the kernel which should be positive and odd. We also should specify the standard deviation in the X and Y directions, sigmaX and sigmaY respectively. If only sigmaX is specified, sigmaY is taken as the same as sigmaX. If both are given as zeros, they are calculated from the kernel size. Gaussian blurring is highly effective in removing Gaussian noise from an image.

```python
img = cv.imread('images/istanbul.jpg')

gaussian_blur = cv.GaussianBlur(img, (5,5), 0)
cv.imshow('Gaussian Blur', gaussian_blur)

plt.imshow(gaussian_blur)
cv.waitKey(0)
```

##### Median

Here, the function cv.medianBlur() takes the median of all the pixels under the kernel area and the central element is replaced with this median value. This is highly effective against salt-and-pepper noise in an image. Interestingly, in the above filters, the central element is a newly calculated value which may be a pixel value in the image or a new value. But in median blurring, the central element is always replaced by some pixel value in the image. It reduces the noise effectively. Its kernel size should be a positive odd integer.

```python
img = cv.imread('images/istanbul.jpg')

median_blur = cv.medianBlur(img,5)
cv.imshow('Median Blur', median_blur)

plt.imshow(median_blur)
cv.waitKey(0)
```

### Canny

OpenCV puts all the above in single function, cv.Canny(). We will see how to use it. First argument is our input image. Second and third arguments are our minVal and maxVal respectively. Third argument is aperture_size. It is the size of Sobel kernel used for find image gradients. 

```python
img = cv.imread('images/istanbul.jpg')

canny = cv.Canny(img, 100, 150)
cv.imshow('Canny', canny)

plt.imshow(canny)
cv.waitKey(0)
```

### Dilation

It is just opposite of erosion. Here, a pixel element is 1 if atleast one pixel under the kernel is 1. So it increases the white region in the image or size of foreground object increases. Normally, in cases like noise removal, erosion is followed by dilation. Because, erosion removes white noises, but it also shrinks our object. So we dilate it. Since noise is gone, they wont come back, but our object area increases. It is also useful in joining broken parts of an object.

```python
img = cv.imread('images/istanbul.jpg')

dilated = cv.dilate(img, (7,7), iterations=3)
cv.imshow('Dilated', dilated)

plt.imshow(dilated)
cv.waitKey(0)
```

### Erosion

The basic idea of erosion is just like soil erosion only, it erodes away the boundaries of foreground object (Always try to keep foreground in white). So what does it do? The kernel slides through the image (as in 2D convolution). A pixel in the original image (either 1 or 0) will be considered 1 only if all the pixels under the kernel is 1, otherwise it is eroded (made to zero).

```python
img = cv.imread('images/istanbul.jpg')

eroded = cv.erode(img, (7,7), iterations=3)
cv.imshow('Eroded', eroded)

plt.imshow(eroded)
cv.waitKey(0)
```
