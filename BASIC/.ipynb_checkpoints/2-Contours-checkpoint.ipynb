```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt

img = cv.imread('images/hand.jpg')
```

### 1 - Convert Image to Gray

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 

plt.imshow(gray)
```

###  2 - Find Edges with Canny

```python
edged = cv2.Canny(gray, 30, 200)

plt.imshow(edged)
```

### 3 - Finding Contours 

```python
contours, hierarchy = cv2.findContours(edged,  cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
```

### 4 - Contours Length

```python
print(f'{len(contours)} contour(s) found!')
```

### 5 - Draw Contours

```python
# All contours are drawn with -1
# source, contours, all draw, color-code, thickness
cv2.drawContours(image, contours, -1, (0, 255, 0), 5) 
cv2.imshow('Contours', image) 

plt.imshow(image)
cv2.waitKey(0) 
cv2.destroyAllWindows() 
```
