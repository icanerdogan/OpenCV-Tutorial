```python
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread('images/istanbul.jpg')

gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv.imshow('Gray', gray)
plt.imshow(gray)
```

### Simple Thresholding

Here, the matter is straight forward. If pixel value is greater than a threshold value, it is assigned one value (may be white), else it is assigned another value (may be black). The function used is cv2.threshold. First argument is the source image, which should be a grayscale image. Second argument is the threshold value which is used to classify the pixel values. Third argument is the maxVal which represents the value to be given if pixel value is more than (sometimes less than) the threshold value. 

OpenCV provides different styles of thresholding and it is decided by the fourth parameter of the function. Different types are:

##### THRESH_BINARY

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

threshold, thresh_binary = cv.threshold(gray, 100, 300, cv.THRESH_BINARY)
cv.imshow('Simple Thresh Binary', thresh_binary)

plt.imshow(thresh_binary)

cv.waitKey(0)
```

##### THRESH_BINARY_INV

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

threshold, thresh_inverse = cv.threshold(gray, 100, 300, cv.THRESH_BINARY_INV )
cv.imshow('Simple Thresh Binary Inverse', thresh_inverse)

plt.imshow(thresh_inverse)

cv.waitKey(0)
```

##### THRESH_TRUNC

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

threshold, thresh_trunc = cv.threshold(gray, 100, 300, cv.THRESH_TRUNC )
cv.imshow('Simple Thresh Trunc', thresh_trunc)

plt.imshow(thresh_trunc)

cv.waitKey(0)
```

##### THRESH_TOZERO

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

threshold, thresh_tozero = cv.threshold(gray, 100, 300, cv.THRESH_TOZERO)
cv.imshow('Simple Thresh Tozero', thresh_tozero)

plt.imshow(thresh_tozero)

cv.waitKey(0)
```

##### THRESH_TOZERO_INV

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

threshold, thresh_tozero_inverse = cv.threshold(gray, 100, 300, cv.THRESH_TOZERO_INV)
cv.imshow('Simple Thresh Inverse', thresh_tozero_inverse)

plt.imshow(thresh_tozero_inverse)

cv.waitKey(0)
```

### Adaptive Thresholding

In the previous section, we used a global value as threshold value. But it may not be good in all the conditions where image has different lighting conditions in different areas. In that case, we go for adaptive thresholding. In this, the algorithm calculate the threshold for a small regions of the image. So we get different thresholds for different regions of the same image and it gives us better results for images with varying illumination.

##### ADAPTIVE_THRESH_GAUSSIAN_C

Threshold value is the weighted sum of neighbourhood values where weights are a gaussian window.

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

adaptive_gaussian = cv.adaptiveThreshold(gray, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY_INV, 11, 2)
cv.imshow('Adaptive Thresh Gaussian', adaptive_gaussian)

plt.imshow(adaptive_gaussian)

cv.waitKey(0)
```

##### ADAPTIVE_THRESH_MEAN_C

Threshold value is the mean of neighbourhood area.

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

adaptive_mean = cv.adaptiveThreshold(gray, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY_INV, 11, 2)
cv.imshow('Adaptive Thresh Mean', adaptive_mean)

plt.imshow(adaptive_mean)

cv.waitKey(0)
```
