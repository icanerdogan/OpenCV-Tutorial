```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

### Blank Window

```python
blank = np.zeros((600,600, 3), dtype='uint8')

plt.imshow(blank)
```

### Coordinate Draw

```python
# y-start:y-stop , x-start:x-stop
blank[100:500, 200:300] = 255, 0, 0

cv.imshow('Red Rectangle', blank)

plt.imshow(blank)

cv.waitKey(0)
```

### Rectangle

```python
blank = np.zeros((600,900, 3), dtype='uint8')

# page, start-coordinate, (width, height), color-code, thickness
# thickness is -1, the inside of the shape is painted
cv.rectangle(blank, (0,0), (blank.shape[1]//2, blank.shape[0]//2), (0,255,0), thickness=-1) 
cv.imshow('Rectangle', blank)

plt.imshow(blank)

cv.waitKey(0)
```

### Circle

```python
blank = np.zeros((600,600, 3), dtype='uint8')

# page, (center-coordinate), radius, color-code, thickness
cv.circle(blank, (blank.shape[1]//2, blank.shape[0]//2), 80, (0,0,255), thickness=-1)
cv.imshow('Circle', blank)

plt.imshow(blank)

cv.waitKey(0)
```

### Line

```python
blank = np.zeros((600,600, 3), dtype='uint8')

# page, start-coordinate, stop-coordinate, color-code, thickness
cv.line(blank, (300,400), (100,200), (255,255,153), thickness=9)
cv.imshow('Line', blank)

plt.imshow(blank)

cv.waitKey(0)
```

### Text

```python
blank = np.zeros((800,1000, 3), dtype='uint8')

# page, text, coordinate, font, font-size, color-code, thickness
cv.putText(blank, 'https://github.com/icanerdogan', (10,400), cv.FONT_HERSHEY_TRIPLEX, 1.7, (255,200,0), 2)
cv.imshow('Text', blank)

plt.imshow(blank)
cv.waitKey(0)
```
